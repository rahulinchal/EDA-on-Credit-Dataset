# Relacing the Null and special character with its respective medians.

```python
data["checking_balance"].replace("unknown", 0, inplace = True)
data["checking_balance"].replace("< 0 DM", 0, inplace = True)
data["checking_balance"].replace("1 - 200 DM", 100, inplace = True)
data["checking_balance"].replace("> 200 DM", 200, inplace = True)

data["savings_balance"].replace("< 100 DM", 100, inplace = True)
data["savings_balance"].replace("unknown", 0, inplace = True)
data["savings_balance"].replace("100 - 500 DM", 300, inplace = True)
data["savings_balance"].replace("500 - 1000 DM", 750, inplace = True)
data["savings_balance"].replace("> 1000 DM", 1000, inplace = True)

data["employment_duration"].replace("1 - 4 years", 2.5, inplace = True)
data["employment_duration"].replace("> 7 years", 7, inplace = True)
data["employment_duration"].replace("4 - 7 years", 5.5, inplace = True)
data["employment_duration"].replace("< 1 year", 1, inplace = True)
data["employment_duration"].replace("unemployed", 0, inplace = True)

data["purpose"].replace("car0", "car", inplace = True)
```

# One hot encoding

```python
dummies=pd.get_dummies(data[['credit_history','purpose','other_credit','housing','job','phone']],
                      columns=['credit_history','purpose','other_credit','housing','job','phone'])
```

```python
dummies.sample(10)
```

# Concatinating the dummies created

```python
data = pd.concat([data, dummies], axis = 1)
data.sample(10)
```

```python
columns = ['credit_history','purpose','other_credit','housing','job','phone']
data.drop(columns, axis= 1, inplace = True)
```

```python
data['savings_balance'].replace('< 100 DM',np.random.randint(1,99),inplace=True)
data.sample(10)
```

```python
def default_new(x):
    if x == "no":
        return 0
    if x == "yes":
        return 1
```

```python
data["default"] = data["default"].apply(default_new)
```
